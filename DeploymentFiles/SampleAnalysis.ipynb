```python
def GetRates(year):
    fx_rates_df = pd.read_csv(f"C:\Work\Programming\Hackathons\Diplomacy software\Historic data\exchange_rates\{year}.csv", index_col=0)
    fx_rates_df.index.name = 'country'

    rates = []

    for country1 in fx_rates_df.index:
        for country2 in fx_rates_df [ fx_rates_df.index == "AUS" ]:
            if country1.strip() != country2.strip():
                rate = fx_rates_df.loc[country1][country2]
                if rate == 1:
                    rate = 0
                rates.append([country2.strip() , country1.strip(), rate])

    return rates # One unit of currency in the first column is equal to the third column units of the currency in the second column
```

```python
RatesYears = []

for i in range(1950, 2023, 5):
    RatesYears.append(GetRates(i))
```

```python
RatesYears[0]
```

```python
rates = []

for country in range(len(RatesYears[0])):

    AvgRateTemp = []

    for year in range(len(RatesYears) - 1):
        try:
            AvgRateTemp.append(
                (RatesYears[year + 1][country][2] - RatesYears[year][country][2]) / RatesYears[year][country][2] * 100
                )
        except TypeError:
            pass
        except ZeroDivisionError:
            pass

    if len(AvgRateTemp) == 0:
        avg = 0
    else:
        try:
            avg = sum(AvgRateTemp) / len(AvgRateTemp)
        except TypeError:
            Total = []

            for i in AvgRateTemp:
                if isinstance(i, int):
                    Total.append(i)

            if len(AvgRateTemp) != 0:
                avg = sum(Total) / len(Total)
            else:
                avg = 0

    rates.append( [RatesYears[0][country][0],
                   RatesYears[0][country][1],
                   avg ] )
```

```python
def findold(combo, data):
    # Combo is a list of length 2

    indices = []

    for i in range(len(data)):
        try:
            if data[i][0] == combo[0]:
                indices.append(i)
        except TypeError:
            return -1

    for ind in indices:
        if data[ind][1] == combo[1]:
            return ind # The positon of combo in data
        
    return -1
```

```python
def find(combo, data):
    # Combo is a list of length 2

    combo = combo.sort()

    indices = []

    for i in range(len(data)):
        try:
            if data[i][0] == combo[0]:
                indices.append(i)
        except TypeError:
            return -1

    for ind in indices:
        if data[ind][1] == combo[1]:
            return ind # The positon of combo in data
        
    return -1
```

```python
fdi_flows = [] # Going from country index 0 to country index 1
```

```python
for i in fdi_flows_df.index:
    if fdi_flows_df['Flow direction'][i] == "In":
        fwd = False
    else:
        fwd = True
       
    if float(fdi_flows_df[' Value'][i]) < 0:
        fwd = not fwd

    if fwd:
        ind = findold( [fdi_flows_df[' Reporting country'][i], fdi_flows_df[' Partner country'][i]], fdi_flows)
    else:
        ind = findold( [fdi_flows_df[' Partner country'][i], fdi_flows_df[' Reporting country'][i]], fdi_flows)
    
    if ind >= 0:
        fdi_flows[ind][2].append(fdi_flows_df[' Value'][i])
    else:
        if fwd:
            fdi_flows.append( [ fdi_flows_df[' Reporting country'][i] , fdi_flows_df[' Partner country'][i] , [ fdi_flows_df[' Value'][i] ] , fdi_flows_df[' Currency'][i]] )
        else:
            fdi_flows.append( [ fdi_flows_df[' Partner country'][i] , fdi_flows_df[' Reporting country'][i] , [ fdi_flows_df[' Value'][i] ] , fdi_flows_df[' Currency'][i]] )

# Takes 2 mins to run
```

**Princple of computation**

The scores are out of 10 for each of the two metrics.

For direct relation, find the percentile
For an indirect relation, find 100 - percentile

```python
def Get_Percentile(value, column):
    column = column.sort_values()
    column = list(column)
    pos = column.index(value)
    return pos / len(column)
```

```python
resilience = []
SurvivalInstinct = []

for country in NamedCountries:
    # Get resilience score
    print(country)
    row = master_df.loc[master_df['Country'] == country ]
    no1 = Get_Percentile(float(row['GDP Per Capita']), master_df['GDP Per Capita'])
    no2 = Get_Percentile(float(row['HDI']), master_df['HDI'])
    no3 = Get_Percentile(float(row['Life expectancy']), master_df['Life expectancy'])
    no4 = Get_Percentile(float(row['Expected years of schooling']), master_df['Expected years of schooling'])
    no5inv = Get_Percentile(float(row['Gini']), master_df['Gini'])

    res_Score = ( no1 + no2 + no3 + no4 + (1 - no5inv) ) / 5 * 10

    resilience.append(res_Score)

    no1 = Get_Percentile(float(row['GDP']), master_df['GDP'])
    no2 = Get_Percentile(float(row['GNI']), master_df['GNI'])
    no3 = Get_Percentile(float(row['Military']), master_df['Military'])
    no4 = Get_Percentile(float(row['FDI inflow']), master_df['FDI inflow'])

    surv_score = ( no1 + no2 + no3 + no4 ) / 4 * 10

    SurvivalInstinct.append(surv_score)
```

## Bilateral

```python
# Normalise all the values

def Normalise(data):
    vals = []

    for i in range(len(data)):
        try:
            val = int(data[i][2])
            vals.append(data[i][2])
        except ValueError:
            vals.append(0)
        except OverflowError:
            vals.append(0)

    mean = statistics.mean(vals)
    std = statistics.pstdev(vals)

    for i in range(len(data)):
        data[i][2] = ( data[i][2] - mean ) / std

    return data

migration_flows = Normalise(migration_flows)
fdi_flows = Normalise(fdi_flows)
rates = Normalise(rates)
```

# Final weights

```python
weight_adjust = []

for i in range(len(NamedCountries)):
    for j in range( len(NamedCountries) - i -1 ):
        combo =  [NamedCountries[i], NamedCountries[j]]

        pos = find(combo, migration_flows)

        migration = migration_flows[pos][2]

        pos = find(combo, fdi_flows)

        fdi = fdi_flows[pos][2]

        pos = find(combo, rates)

        rate = rates[pos][2]

        Disagreement = ( - 1 ) * fdi
        Bond = migration - rate
        Geography = ( geodesic(Position[i], Position[j]).kilometers - 8259) / 4558

        Weight = Disagreement - Bond - Geography + 1 # The + 1 adjusts for two negatives as opposed to one positive

        Weight = math.exp(Weight)

        res = ( monolateral_data[NamedCountries[i]][0] + monolateral_data[NamedCountries[j]][0] ) / 2
        Survival = ( monolateral_data[NamedCountries[i]][1] + monolateral_data[NamedCountries[j]][1] ) / 2

        Weight = Weight * ( 10 - res ) / 5
        Weight = Weight * ( 10 - Survival ) / 5

        weight_adjust.append( [ NamedCountries[i], NamedCountries[j], Weight] )
        # Target to finish by today (12th January)
```

```python
for i in range(len(NamedCountries)):
    for j in range(len(NamedCountries)):
        if i != j:
            country = NamedCountries[i]
            country2 = NamedCountries[j]

            lloc = "Weights\\" + country + ".bin"

            with open(lloc, "rb") as f:
                weights = pickle.load(f) 

            ind = find( [country, country2], weight_adjust)    

            weights[country2] = [[random.gauss( weight_adjust[ind][2] , 0.1) for i in range(9)] for j in range(7)]

            with open(lloc, "wb") as f:
                pickle.dump(weights, f)

            print(f"Completed {country} and {country1}.")
```
