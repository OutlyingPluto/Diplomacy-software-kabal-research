Create parallel arrays where the data about each country is stored

GDP

```python
GDP_df = pd.read_csv("C:\Work\Programming\Hackathons\Diplomacy software\Historic data\gdp.csv")
```

```python
def GetGDP(country):    
    row = GDP_df.loc[GDP_df['Country Name'] == country ]

    year = 2022

    GDP = 0

    while GDP == 0:
        print(row[f' {year} [YR{year}]'])
        GDP = int(row[f' {year} [YR{year}]'])
        year -= 1

    return GDP
```

```python
GDP_data = []

for country in NamedCountries:
    try:
        GDP_data.append(GetGDP(country))
    except:
        GDP_data.append(0)
```
