# Combine metrics

**Princple of computation**

The scores are out of 10 for each of the two metrics.

For direct relation, find the percentile
For an indirect relation, find 100 - percentile

```python
master_df = pd.DataFrame(list(zip(NamedCountries, GDP_data, GDP_PerC_data, GNI_data, Military_data, Gini_data, fdi_inflow_data, fdi_outflow_data, HDI_data, Life_data, School_data)), columns=['Country', 'GDP', 'GDP Per Capita', 'GNI', 'Military', 'Gini', 'FDI inflow', 'FDI outflow', 'HDI', 'Life expectancy', 'Expected years of schooling'])
```

```python
master_df
```

```python
col = master_df['Life expectancy']
col = col.to_numpy()
row = master_df.loc[master_df['Country'] == NamedCountries[0] ]
np.where(col == float(row['Life expectancy']))
```

```python
row['Life expectancy']
```

```python
def Get_Percentile(value, column):
    column = column.sort_values()
    column = list(column)
    pos = column.index(value)
    return pos / len(column)
```

```python
Get_Percentile(4232173916086, master_df['GDP'])
```

```python
row = master_df.loc[master_df['Country'] == NamedCountries[1] ]
int(row['GDP Per Capita'])
```

```python
resilience = []
SurvivalInstinct = []

for country in NamedCountries:
    # Get resilience score
    print(country)
    row = master_df.loc[master_df['Country'] == country ]
    no1 = Get_Percentile(float(row['GDP Per Capita']), master_df['GDP Per Capita'])
    no2 = Get_Percentile(float(row['HDI']), master_df['HDI'])
    no3 = Get_Percentile(float(row['Life expectancy']), master_df['Life expectancy'])
    no4 = Get_Percentile(float(row['Expected years of schooling']), master_df['Expected years of schooling'])
    no5inv = Get_Percentile(float(row['Gini']), master_df['Gini'])

    res_Score = ( no1 + no2 + no3 + no4 + (1 - no5inv) ) / 5 * 10

    resilience.append(res_Score)

    no1 = Get_Percentile(float(row['GDP']), master_df['GDP'])
    no2 = Get_Percentile(float(row['GNI']), master_df['GNI'])
    no3 = Get_Percentile(float(row['Military']), master_df['Military'])
    no4 = Get_Percentile(float(row['FDI inflow']), master_df['FDI inflow'])

    surv_score = ( no1 + no2 + no3 + no4 ) / 4 * 10

    SurvivalInstinct.append(surv_score)
```

```python
print(resilience)
print(SurvivalInstinct)
```
